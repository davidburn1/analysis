```python
%matplotlib inline  
import numpy as np
import matplotlib.pyplot as plt

import sys
sys.path.append('../../')
from msgAnalysis.src.xmcd import loadXMCD

plt.rcParams.update({'font.size':14, 'xtick.top':True, 'ytick.right':True, 'xtick.direction':'in', 'ytick.direction':'in'})
```

```python
directory="./example_data/xmcd_data"
```

```python
#Fe
data = loadXMCD(directory, [620515], [620516] )

fig, ax = plt.subplots(2, 1, figsize=(6, 5), facecolor='w', sharex=True )

ax[0].plot(data['energy'], data['m17']['a'], 'r', label='a')
ax[0].plot(data['energy'], data['m17']['b'], 'b', label='b')
ax[1].plot(data['energy'], data['m17']['a']-data['m17']['b'], 'k')

ax[0].legend()
ax[0].set_title("Fe")
ax[1].set_xlabel("Energy (eV)")
ax[0].set_ylabel("XAS")
ax[1].set_ylabel("Difference")
plt.subplots_adjust( hspace=0.05, wspace=0.05)
plt.show()

print("Minimum at: %2.2f eV" % data['m17']['min'])
```
