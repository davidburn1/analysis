```python
%matplotlib inline  
import numpy as np
import matplotlib.pyplot as plt

import sys
sys.path.append('../src')
from poms import loadSampleAlignmentData

plt.rcParams.update({'font.size': 14, 'xtick.top': True, 'ytick.right': True, 'xtick.direction': 'in', 'ytick.direction': 'in'})
```

Collect data using flying scans on the user1_axis2 axis at several different user1_axis1 positions. Use this example to construct a colormap of the sample intensity as a function of position.

```python
directory="./example_data/sample_alignment_data"
```

```python
scans = np.arange(607844, 607851, 1)

ax1, ax2, refl = loadSampleAlignmentData(directory, scans)

plt.tripcolor(ax2,ax1,  refl.T, cmap="jet", shading="flat")

plt.xlabel("Axis 2")
plt.ylabel("axis 1")
plt.show()
```
